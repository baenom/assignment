```python
# 병렬 퍼셉트론 코드
import random
import math

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

def perceptron(x,count,w,b,correct):
    z = x * w[count] + b[count]
    answer = sigmoid(z)
    learning = 0.01
    w_l = w[count]+(learning*(correct[count]-answer)*x)
    b_l = b[count]+(correct[count]-answer)
    w[count] = w_l
    b[count] = b_l
    if count+1 == len(w):
        return answer,w,b
    count += 1
    return perceptron(x,count,w,b,correct)
x = []
y = []
user_input = input("데이터")
data = list(map(int, user_input.split()))
x = data
user_input = input("답")
data = list(map(int, user_input.split()))
count = 0
correct = data
correct_list_out = []



w_list = [random.uniform(0, 1) for _ in range(len(correct))]
b_list = [random.uniform(0, 1) for _ in range(len(correct))]

for i in range(len(x)):
    if i == 0:
        correct_list_hidden = perceptron(x[i],0,w_list,b_list,correct)
    else:
        correct_list_hidden = perceptron(x[i],0,correct_list_hidden[1],correct_list_hidden[2],correct)
    print(x[i],correct_list_hidden[0])
    correct_list_out.append(correct_list_hidden[0])

print(correct_list_out)
```

```python
# 신경망코드
import random
import math

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

def perceptron(x,count,w,b,correct):
    z = x * w[count] + b[count]
    answer = sigmoid(z)
    learning = 0.01
    w_l = w[count]+(learning*(correct[count]-answer)*x)
    b_l = b[count]+(correct[count]-answer)
    w[count] = w_l
    b[count] = b_l
    if count+1 == len(w):
        return answer,w,b
    count += 1
    return perceptron(x,count,w,b,correct)
x = []
y = []
user_input = input("데이터")
data = list(map(int, user_input.split()))
x = data
user_input = input("답")
data = list(map(int, user_input.split()))
count = 0 
count_layer = 0

correct = data


layers_w = [[random.uniform(0, 1) for _ in range(len(correct))] for _ in range(len(correct))]
layers_b = [[random.uniform(0, 1) for _ in range(len(correct))] for _ in range(len(correct))]

def layer(x,count_layer,correct_list_out):
    if count_layer >= len(layers_w):
        return correct_list_out
    
    w_list = layers_w[count_layer]
    b_list = layers_b[count_layer]
    for i in range(len(x)):
        if i == 0:
            correct_list_hidden = perceptron(x[i],0,w_list,b_list,correct)
        else:
            correct_list_hidden = perceptron(x[i],0,correct_list_hidden[1],correct_list_hidden[2],correct)
        correct_list_out.append(correct_list_hidden[0])

    layers_w[count_layer] = correct_list_hidden[1]
    layers_b[count_layer] = correct_list_hidden[2]
    
    return layer(correct_list_out,count_layer+1,correct_list_out)
    

print(layer(x,count_layer,list())[-len(correct):])
```
